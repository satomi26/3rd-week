```python
df_one = df.drop(["Summary"], axis = 1)
df_one
```

```python
df_two = df_one.rename(columns={"MovieID": "movie_id", "Title": "title","MPAA Rating": "mpaa_rating","Budget": "budget", "Gross": "gross", "Release Date": "release_date","Genre": "genre","Runtime": "runtime", "Rating":"rating", "Rating Count":"rating_count"})
df_two
```

```python
df_three = df_two[:-1]
df_three
```

```python
df_three.isna().sum()
```

```python
# we have missing data in column "rating" and "rating count"
# find the mean and median from "rating" and "rating count"

df_mean = df_three.mean()
df_mean
```

```python
# use fillna() method to replace the missing values of Rating with their median

df_three["rating"] = df_three["rating"].fillna(value = df_three["rating"].mean())


df_three
```

```python
df_three["rating_count"] = df_three["rating_count"].fillna(value = df_three["rating_count"].mean())
df_three
```

```python
df_three.shape
```

```python
df_three.isna().sum()
```

```python
df_four = df_three.drop_duplicates()
df_four
```

```python
duplicateRows = df_three[df_three.duplicated()]
duplicateRows
```

```python
# This is where we used a parameter keep = false to show all duplicates

duplicateRows_title = df_three[df_three.duplicated(['title'], keep = False)]
duplicateRows_title
```

```python
without_duplicate = duplicateRows_title.drop_duplicates()
without_duplicate
```

```python
# Convert Budget into integer type (use astype() method)
df_four = df_three.astype({'budget':'int'})

df_four
```

```python
df_four.dtypes
```

```python
# convert "gross" to integer

df_five = df_four.astype({'gross':'int'})
df_five
```

```python
df_five.dtypes
```

```python
# Convert Release Date into date type (use to_datetime method)

# df_six = df_five.to_datetime({"release_date" : ""})

df_five["release_date"] = pd.to_datetime(df_five["release_date"], format = "%Y/%m/%d")
```

```python
df_five.dtypes
```

```python
cleaned_df = df_five
cleaned_df
```

```python
# most & least frequent MPAA Rating in the dataset in terms of occurances (発生頻度からいうと)

frequent_mpaa = cleaned_df["mpaa_rating"].value_counts()
print(frequent_mpaa)
print(f"most frequent MPAA Rating is: ", frequent_mpaa.idxmax())
print(f"least frequent MPAA Rating is: ", frequent_mpaa.idxmin())
```

```python
frequent_mpaa = cleaned_df["mpaa_rating"].value_counts()
print(frequent_mpaa)
print(f"most frequent MPAA Rating is: ", frequent_mpaa.index[0])
```

```python
# most and least expensive Genre (take an average of all Budget measures grouped by Genre - use groupBy() method)

average_budget = cleaned_df.groupby("genre").mean().reset_index()

average_budget[['gross', 'budget']] = average_budget[['gross', 'budget']].round()

average_budget.sort_values("budget")

# most expensive genre : Fantasy
# least expensive genre: Horror
```
